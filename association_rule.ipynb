## Association Rule Mining:

```python
!pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
te = TransactionEncoder()
```

```python
transactions = data.apply(lambda x: x[0].split(','), axis = 1).tolist()
```

```python
te_array = te.fit_transform(transactions)
```

```python
te_array
```

```python
df = pd.DataFrame(te_array, columns = te.columns_)
```

```python
df
```

```python
te.columns_
```

```python
frequent_itemsets = apriori(df, min_support = 0.02, use_colnames = True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, num_itemsets = 5176, min_threshold = 0.4)
```

```python
rules
```

```python
rel_rules = rules[rules.lift>1.5]
rel_rules
```

## Interview Questions:

###  1. What is Lift and Why is it Important in Association Rules?
#### Lift measures how much more likely two items are to be bought together compared to if they were bought independently. 
#### It helps identify strong relationships between products.
#### Lift > 1: Items are more likely to be bought together than by chance (strong association).
#### Lift = 1: No relationship. items are bought independently.
#### Lift < 1: Items are less likely to be bought together than expected by chance.

### 2. What are Support and Confidence? How Do You Calculate Them?
#### Support: Measures how frequently an itemset appears in the dataset.
#### Confidence: Measures how often an item Y appears in transactions that contain item X.

### 3. What Are Some Limitations or Challenges of Association Rule Mining?
#### High Computational Cost: Analyzing large datasets can be slow and resource-intensive.
#### Too Many Rules: Generates many rules, making it hard to find the most useful ones.
#### Low Interpretability: Some rules may be statistically valid but not practically useful.
#### Rare Item Problem: Items with low frequency may not appear in rules despite their importance.
#### Redundant Rules: Many rules may be similar, providing little new insight.
